```python
import pandas as pd
netflix_df =  pd.read_csv('/content/drive/MyDrive/netflix_titles.csv')

netflix_df.head(5)
```

```python
netflix_df['date_added'] = pd.to_datetime(netflix_df['date_added'], errors='coerce')
netflix_df.head(5)
```

```python

netflix_df['director'].fillna('Not Available', inplace=True)
netflix_df['cast'].fillna('Not Available', inplace=True)
netflix_df['country'].fillna('Unknown', inplace=True)
netflix_df['rating'].fillna('Not Rated', inplace=True)

```

```python
netflix_df.isnull().sum()
```

```python

netflix_df['year_added'] = netflix_df['date_added'].dt.year
netflix_df['month_added'] = netflix_df['date_added'].dt.month
netflix_df.head(5)
```

```python

netflix_df[['duration_int', 'duration_type']] = netflix_df['duration'].str.extract(r'(\d+)\s*(\w+)')
netflix_df['duration_int'] = pd.to_numeric(netflix_df['duration_int'])

netflix_df.head(5)
```

#Content Type Breakdown (Movies vs TV Shows)



```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(6,4))
sns.countplot(data=netflix_df, x='type', palette='pastel')
plt.title('Content Type Distribution')
plt.xlabel('Type')
plt.ylabel('Count')
plt.show()
```

```python
type_counts = netflix_df['type'].value_counts()

plt.figure(figsize=(6,6))
plt.pie(type_counts, labels=type_counts.index, autopct='%1.1f%%', startangle=140, colors=['#FF9999','#66B3FF'])
plt.title('Proportion of Movies vs TV Shows')
plt.axis('equal')
plt.show()
```

#Titles Added Per Year



```python
plt.figure(figsize=(10,5))
netflix_df['year_added'].value_counts().sort_index().plot(kind='bar', color='skyblue')
plt.title('Number of Titles Added Each Year')
plt.xlabel('Year Added')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.show()
```

```python
titles_per_year = netflix_df['year_added'].value_counts().sort_index()

plt.figure(figsize=(10,5))
plt.plot(titles_per_year.index, titles_per_year.values, marker='o', color='darkblue')
plt.title('Trend of Titles Added Over the Years')
plt.xlabel('Year Added')
plt.ylabel('Number of Titles')
plt.grid(True)
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(10,5))
netflix_df['country'].value_counts().head(10).plot(kind='bar', color='lightcoral')
plt.title('Top 10 Countries with Most Content')
plt.xlabel('Country')
plt.ylabel('Number of Titles')
plt.xticks(rotation=45)
plt.show()
```

```python

netflix_df_exploded = netflix_df.explode('genres')

plt.figure(figsize=(10,5))
netflix_df_exploded['genres'].value_counts().head(10).plot(kind='bar', color='mediumseagreen')
plt.title('Top 10 Most Common Genres')
plt.xlabel('Genre')
plt.ylabel('Number of Titles')
plt.xticks(rotation=45)
plt.show()
```

```python
monthly = netflix_df['month_added'].value_counts().sort_index()

plt.figure(figsize=(8,4))
monthly.plot(kind='bar', color='goldenrod')
plt.title('Titles Added by Month')
plt.xlabel('Month')
plt.ylabel('Number of Titles')
plt.xticks(rotation=0)
plt.show()
```

```python
plt.figure(figsize=(10,5))
sns.histplot(data=netflix_df[netflix_df['type'] == 'TV Show'], x='duration_int', bins=15, color='orchid')
plt.title('Distribution of TV Show Seasons')
plt.xlabel('Number of Seasons')
plt.ylabel('Number of Shows')
plt.show()
```
